### Face Recognition System Challenges

### Despite the fact that face recognition has been a mature research area, there are still many problems that must be overcome in order to develop a robust face recognition system that works well under various conditions such as illumination, pose, expressions, illumination and expressions, illumination and pose, and finally illumination and expression and pose variations, as shown in the below image. The results show that all of the recognition approaches performed well on huge face databases captured in well-controlled conditions. However, in uncontrolled conditions, their performance suffers mostly as a result of fluctuations in lighting and head rotations. Such variances have proven to be one of the most serious issues with facial recognition systems.



### These approaches may fail to effectively portray the faces because to substantial changes in lighting conditions, facial expression, and other reasons. The primary reason for this is that the face patterns are located on a complicated nonlinear and nonconvex manifold in highdimensional space.
### Nonlinear extensions such as kernel PCA (KPCA), kernel LDA (KLDA) or locally linear embedding (LLE) have been proposed to cope with such instances. The most nonlinear approaches use kernel techniques, where the fundamental concept is to translate the input face pictures into a higherdimensional space where the face manifold is linear and simplified. As a result, classic linear approaches can be used.
### Although PCA, LDA, and LRC are all considered linear subspace learning algorithms, it is worth noting that PCA and LDA approaches focus on the global structure of the Euclidean space, whereas the LRC approach focuses on the manifold's local structure.
### Face is projected into a linear subspace covered by the eigenface images in these approaches. The distance in face space is the distance along the plane from the mean picture, whereas the distance from face space is the orthogonal distance to the plane. Both of these distances may be converted to Mahalanobis distances and assigned probabilistic interpretation.
### Despite the solid theoretical underpinning of kernelbased approaches, their actual use in face recognition issues does not yield a substantial advantage over linear methods.
### Another nonlinear projection technique family has been developed. They inherited the linear techniques' simplicity and the nonlinear approaches' capacity to cope with complicated data. Among these approaches, LLE  and locality preserving projection (LPP) stand out. They generate a projection scheme for training data alone, but its ability to project fresh data items is dubious.

## Question 5. Use an image from folder Step 5, then, write a code to detect faces in the image, print the number of detected faces on the screen and finally, draw a rectangle around their faces. 

## Answer:- Face Detection

```python
# Loading all the necessary libraries for Face Detection

import PIL.Image      # it's python image library 
import PIL.ImageDraw   # PIL.Imagedraw is used to perform drawings on images
import face_recognition   # this library will help recognizing face and performing all operation related to face recognition
import glob           # This library will help fetching files from system
```

```python
# Detecting faces in image


# Load the jpg file into a numpy array
image = face_recognition.load_image_file("face_recognition/Application Exercise Images/Step 5/image1.jpeg")

# Find all the faces in the image
face_locations = face_recognition.face_locations(image)
number_of_faces = len(face_locations)
print(f"Total {number_of_faces} face(s) found in this photograph.")


# Load the image into a Python Image Library object 
# so that we can draw on top of it and display it

pil_image = PIL.Image.fromarray(image)

for i in face_locations:

    top, right, bottom, left = i

    # Let's draw a box around the face in the image
    draw = PIL.ImageDraw.Draw(pil_image)
    draw.rectangle([left, top, right, bottom], outline="red", width = 3)

# Display the image on screen
pil_image.show()
```

## Output :

### In the above picture we can see that we detected each face and highlites it with red rectangle.

## Question 6. Then, extend the code from Step 5 such that it detects individual landmarks in every face in the image and draws lines representing those landmarks.

## Answer:- Face Landmarks Detection

```python
# Let's Find all facial features in all the faces which were detected in above image
face_landmarks_list = face_recognition.face_landmarks(image)

# Load the image into a Python Image Library again to remove square that we drawn in previous code
# so we can have better visibility of all landmark elements of the face.

pil_image = PIL.Image.fromarray(image)

# Create a PIL drawing object to be able to draw lines on all facial landmarks
draw = PIL.ImageDraw.Draw(pil_image)

# Loop over each face to detect facial landmarks
for i in face_landmarks_list:

    # Loop over each facial feature like "eye", "nose", "mouth", "lips", etc
    for name, list_of_points in i.items():
        # Let's trace out each facial feature in the image with a line!
        draw.line(list_of_points, fill= "blue" , width=2)

# Display image with highlited facial features
pil_image.show()
```

## Output :

### In above Image we can see that we successfully completed task of highliting facial landmarks like "Eyes",  "Nose", "Nose tip", "Chin", "Eyebrows" and "Lips".

## Question 7. In this step, you must use the provided sample dataset and known image in folder Step 7 to code a face recognition system. The system must be able to find a match in the dataset for John.jpg and print out how many times John gets a match in the dataset.

## Answer:- How many times John gets a match in the dataset.

```python
# Load the known images to encode it and compare it with unknown images
john = face_recognition.load_image_file("face_recognition/Application Exercise Images/john.jpg")

# Get the face encoding of john.
john_encoding = face_recognition.face_encodings(john)

# Load all the unknow images to check with john's face
file = "C:/Users/Darshan/Desktop/SEM-2 AI and DS/ML/face_recognition/Application Exercise Images/Step 7/*.jpg"
unknown_images = [face_recognition.load_image_file(image) for image in glob.glob(file)]
unknown_face_encodings = []

# Get face encodings for any people in the unknown image
for j in unknown_images :
    a = face_recognition.face_encodings(j)
    unknown_face_encodings.append(a)

count = 0

# There might be more than one person in the image, so we need to loop over each face we found

for unknown_face_encoding in unknown_face_encodings:
    
    for face in unknown_face_encoding :

        # Check if this unknown face encoding matches with john's encoding.
        detected = face_recognition.compare_faces(john_encoding, face, tolerance=0.6)
    
        if detected[0] :  
            count += 1
            
print (f"john get {count} matches out of {len(unknown_face_encodings)} images.")
```
